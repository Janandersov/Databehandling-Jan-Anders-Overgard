### Pandas Timeseries
A series or dataframe using a datetime index instead of a range index.

Used for storing events/data that fits on a timeline.
- Weather data.
- Temperature readings.
- Heart rate monitoring (EKG).
- Quarterly sales.
- Stock prices.

### Resampling
Resampling is the practice of creating new samples with a lower or higher frequency than the original data.


#### Downsampling
When the sample frequency is lower than the original we can aggregate values from the original data.

```python
df
```

```python
df.resample("2D")
```

```python
from helpers import hdisplay

resample_method = "2D"

hdisplay([
    df.head(10),
    df.resample(resample_method).agg({"n": "max", "rand": "sum"}).head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "W"

hdisplay([
    df.head(10),
    df.resample(resample_method).agg({"n": "max", "rand": "sum"}).head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "M"

hdisplay([
    df.head(10),
    df.resample(resample_method).agg({"n": "max", "rand": "sum"}).head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "Y"

hdisplay([
    df.head(10),
    df.resample(resample_method).agg({"n": "max", "rand": "sum"}).head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

#### Upsampling
When the sampling frequency is higher than the original, we can fill between values.
- ffill
- bfill
- nearest
- interpolation
- fillna

```python
resample_method = "2D"

hdisplay([
    df.head(10),
    df.resample(resample_method).agg({"n": "max", "rand": "sum"}).head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "6H"

hdisplay([
    df.head(10),
    df.resample(resample_method).ffill().head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "6H"

hdisplay([
    df.head(10),
    df.resample(resample_method).bfill().head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "6H"

hdisplay([
    df.head(10),
    df.resample(resample_method).nearest().head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "6H"

hdisplay([
    df.head(10),
    df.resample(resample_method).interpolate().head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "6H"

hdisplay([
    df.head(10),
    df.resample(resample_method).max().fillna(0).head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

```python
resample_method = "6H"

hdisplay([
    df.head(10),
    df.resample(resample_method).max().fillna(df[["n", "rand"]].mean()).head(10)],
    ["Original", f"Resampled using '{resample_method}'"]
)
```

### Working on real data

```python
apple = pd.read_csv("../Data/HistoricalData_Apple.csv", parse_dates=["Date"], index_col="Date")
```

```python
apple.rename(columns={"Close/Last": "Close"}, inplace=True)
apple.info()
```

```python
apple.head()
```

```python
apple[["Close", "Open", "High", "Low"]] = apple[["Close", "Open", "High", "Low"]].map(lambda value: value.replace("$", "")).astype("float64")
```

```python
apple.info()
```

```python
apple.head()
```

```python
apple.sort_index(inplace=True)
```

```python
apple
```

```python
apple.loc["2017-04"]
```

```python
apple.loc["2018", "Close"].mean
```

```python
apple.resample("Y").mean()
```

```python
import seaborn as sns
data = apple.loc["2020-10"]
sns.lineplot(data=data, x=data.index, y="Open")
```

```python
data = apple.loc["2020"]
sns.lineplot(data=data, x=data.index, y="Open")
```

```python
data = apple.loc["2020"].resample("M").mean()
sns.lineplot(data=data, x=data.index, y="Open")
```

```python
data = apple.resample("QS").mean()
sns.lineplot(data=data, x=data.index, y="Open")
```

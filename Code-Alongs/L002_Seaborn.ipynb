### Seaborn
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.head(3)
```

```python
tips.tail(7)
```

```python
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", kind="line", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", kind="scatter", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="sex", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", sizes=[10,30,50,70,90,110], data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", row="day", height=2.5, data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", row="time", col="day", height=2.5, data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", col="day", col_wrap=2, height=2.5, data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=tips)
```

```python
sns.catplot(y="day", x="total_bill", kind="strip", data=tips)
```

```python
sns.catplot(y="day", x="total_bill", kind="strip", hue="sex", data=tips)
```

```python
sns.catplot(y="day", x="total_bill", kind="box", hue="sex", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="bar", data=tips)
```

```python
sns.catplot(x="day", kind="count", data=tips)
```

```python
sns.catplot(x="sex", kind="count", hue="sex", data=tips)
```

```python
sns.displot(x="total_bill", data=tips)
```

```python
sns.displot(x="total_bill", bins= 5, data=tips)
```

```python
sns.displot(x="total_bill", bins= 50, data=tips)
```

```python
sns.displot(x="total_bill", kind="kde", data=tips)
```

```python
sns.displot(x="total_bill", kind="kde",bw_adjust=0.25, data=tips)
```
